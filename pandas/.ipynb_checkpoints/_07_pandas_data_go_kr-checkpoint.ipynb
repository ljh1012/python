#### 평당 분양 가격 
- 분양가격 * 3.3

```python
df['평당분양가격'] = df['분양가격']*3.3
df.head(3)
```

```python
df['전용면적'] = df['규모구분'].str.replace('전용면적',"")
df['전용면적'] = df['전용면적'].str.replace('이하',"")
df['전용면적'] = df['전용면적'].str.replace('초과',"-")
df['전용면적'] = df['전용면적'].str.replace(' ',"").str.strip()
```

```python
df['전용면적'].unique()
```

```python
df.head()
```

```python
df = df.drop(['규모구분','분양가격(제곱미터)'], axis=1)
df.head()
```

```python
df.info()
```

```python
df.groupby(['지역명'])['평당분양가격'].mean()
```

```python
df.groupby(['전용면적'])['평당분양가격'].mean()
```

```python
df.groupby(['전용면적',"지역명"])['평당분양가격'].mean()
```

```python
# 지역명, 전용면적으로 평당분양가격의 평균을 구합니다.
df.groupby(["전용면적", "지역명"])["평당분양가격"].mean().unstack().round()
```

```python
# 연도, 지역명으로 평당분양가격의 평균을 구합니다.
g = df.groupby(["연도", "지역명"])["평당분양가격"].mean()
g
```

```python
g.unstack()
```

```python
g.unstack().transpose()
```

```python
import matplotlib.pyplot as plt
plt.rc("font", family="Malgun Gothic") # 한글폰트
```

```python
# 지역명으로 분양가격의 평균을 구하고 선그래프로 시각화 합니다.
g = df.groupby(["지역명"])["평당분양가격"].mean().sort_values(ascending=False)
g.plot()
```

```python
# 지역명으로 분양가격의 평균을 구하고 막대그래프(bar)로 시각화 합니다.
g.plot.bar(rot=0, figsize=(10, 3))
```

```python
# 전용면적으로 분양가격의 평균을 구하고 막대그래프(bar)로 시각화 합니다.
df.groupby(["전용면적"])["평당분양가격"].mean().plot.bar()
```

```python
# 연도별 분양가격의 평균을 구하고 막대그래프(bar)로 시각화 합니다.
# 연도에 소숫점이 생기지 않게 표시하고자 한다면 ax.xaxis.set_major_locator를 사용해서 integer로 설정합니다.
from matplotlib.ticker import MaxNLocator

ax = plt.figure().gca()
df.groupby(["연도"])["평당분양가격"].mean().plot()
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
```

```python
import seaborn as sns
%matplotlib inline
```

```python
# barplot으로 지역별 평당분양가격을 그려봅니다.
plt.figure(figsize=(10, 3))
sns.barplot(data=df, x="지역명", y="평당분양가격", ci=None)
```

```python
# barplot으로 연도별 평당분양가격을 그려봅니다.

sns.barplot(data=df, x="연도", y="평당분양가격")
```

```python
sns.catplot(data=df, x="연도", y="평당분양가격", kind="bar", col="지역명", col_wrap=4)
```

```python
# lineplot으로 연도별 평당분양가격을 그려봅니다.
# hue 옵션을 통해 지역별로 다르게 표시해 봅니다.
plt.figure(figsize=(10, 5)) # plt.figure -> 기본적으로 자동으로 호출, 특별히 조정할 것이 있을 때 사용igsize = (,) -> 그림 크기 조절
sns.lineplot(data=df, x="연도", y="평당분양가격", hue="지역명")
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.) # 범례 추가                                          범례
```
