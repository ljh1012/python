### boolean selection으로 row선택

#### 30대, 1등석에 탄 사람 선택

```python
class_ = df['Pclass'] == 1 #1등석 탄사람
age_ = (df['Age'] >= 30) & (df['Age'] < 40)
print('result:', len(df[class_ & age_]))
df[class_ & age_]
```

#### 여성, 1등석에 탄 사람

```python
class_ = df['Pclass'] == 1 #1등석 탄사람
female_ = df['Sex'] == 'female'
print('result:', len(df[class_ & female_]))
df[class_ & female_]
```

#### 30대, 여성, 1등석, 퀸즈타운 탑승 승객수

```python
class_ = df['Pclass'] == 1                  # 1등석 
age_ = (df['Age'] >= 30) & (df['Age'] < 40) # 30대
female_ = df['Sex'] == 'female'             # 여성
queens_town_ = df['Embarked'] == 'Q'        # QueensTown 

filtered = (class_ & age_ & female_ & queens_town_)

print('result:', len(df[filtered]))
df[filtered]
```

 - 연속적이지 않은 값(대부분의 경우 숫자를 제외한 나머지 값)을 갖는 데이터를 의미

#### Age 변수 변환하기

#### One-hot encoding
 - 범주형 데이터는 분석단계에서 계산이 어렵기 때문에 숫자형으로 변경이 필요함
 - 범주형 데이터의 각 범주(category)를 column레벨로 변경
 - 해당 범주에 해당하면 1, 아니면 0으로 채우는 인코딩 기법
 - pandas.get_dummies 함수 사용
   - drop_first : 첫번째 카테고리 값은 사용하지 않음

```python
pd.get_dummies(df2, columns=['Pclass','Sex','Embarked'], drop_first=False)
```

```python
pd.get_dummies(df2, columns=['Pclass','Sex','Embarked'], drop_first=True)
```
