#### df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False) 
    결측치가 있는 행 or 열을 삭제
    axis : {0: index / 1: columns} 결측치 제거를 진행 할 레이블입니다.
    how : {'any' : 존재하면 제거 / 'all' : 모두 결측치면 제거} 제거할 유형입니다. 포함만 시켜도 제거할지, 
          전무 NA여야 제거할지 정할 수 있습니다.
    tresh : 결측값이 아닌 값이 몇 개 미만일 경우에만 적용시키는 인수 입니다.
    예를들어, tresh값이 3이라면 결측값이 아닌 값이 3개 미만일 경우에만 dropna메서드를 수행합니다.
    subset : dropna메서드를 수행할 레이블을 지정합니다.
    inplace : 원본을 변경할지의 여부입니다.

```python
train2 = train.copy()
```

```python
dic = { 'male' : 0, 'female' : 1}  # 성별(category -> num)
train2['Sex'] = train2['Sex'].map(dic)
train2
```

```python
train2.drop(['Ticket', 'Cabin'], axis = 1, inplace=True)
train2
```

```python
train2.set_index("PassengerId", inplace=True)
train2
```

```python
train2['Prefix'] = train2['Name'].apply(lambda x : x.split(",")[1].split(".")[0].strip())
train2
```

```python
train2['Prefix'].unique()
```

```python
dic = {'Mr' : 0, 'Mrs' : 1, "Miss" : 2, 'Master':3, 'Don':3, 'Rev':3, 'Dr':3, 'Mme':3, 'Ms':3, 'Major':3, 'Lady':3, 'Sir':3, 'Mlle':3, 'Col':3, 'Capt':3, 'the Countess':3,'Jonkheer':3}
train2['Prefix']=train2['Prefix'].map(dic)
train2
```

```python
train2.drop('Name', axis = 1, inplace = True)
```

```python
train2
```

```python
train2.info()
```

```python
prefix_mean = train2.groupby('Prefix')['Age'].mean()
for i in range(len(prefix_mean)): 
    mean = train2[train2['Prefix']==i]['Age'].mean()
    train2.loc[train2['Prefix']==i,'Age'] = train2.loc[train2['Prefix']==i,'Age'].fillna(mean)

train2.info()
```

```python
train2.dropna(inplace=True)
train2.info()
```
